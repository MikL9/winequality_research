```python
import pandas as pd

# Загрузка данных из файлов CSV
red_wine_data = pd.read_csv('winequality/winequality-red.csv', delimiter=';')
white_wine_data = pd.read_csv('winequality/winequality-white.csv', delimiter=';')

red_wine_data
```

```python
white_wine_data
```

```python
# Расчет среднего значения параметра
mean_alcohol_red = red_wine_data['alcohol'].mean()
mean_alcohol_white = white_wine_data['alcohol'].mean()

# Расчет медианы параметра
median_alcohol_red = red_wine_data['alcohol'].median()
median_alcohol_white = white_wine_data['alcohol'].median()

# Расчет экстремумов параметра
min_alcohol_red = red_wine_data['alcohol'].min()
max_alcohol_red = red_wine_data['alcohol'].max()

min_alcohol_white = white_wine_data['alcohol'].min()
max_alcohol_white = white_wine_data['alcohol'].max()

# Вывод результатов
print("Red Wine - Alcohol:")
print("Mean:", mean_alcohol_red)
print("Median:", median_alcohol_red)
print("Min:", min_alcohol_red)
print("Max:", max_alcohol_red)

print("\nWhite Wine - Alcohol:")
print("Mean:", mean_alcohol_white)
print("Median:", median_alcohol_white)
print("Min:", min_alcohol_white)
print("Max:", max_alcohol_white)
```

```python
print("Размер данных до предобработки:")
print("Красное вино:", red_wine_data.shape)
print("Белое вино:", white_wine_data.shape)
```

```python
# Удаление шумов
red_wine_data = red_wine_data[red_wine_data['alcohol'] >= 0]
white_wine_data = white_wine_data[white_wine_data['alcohol'] >= 0]
red_wine_data = red_wine_data[red_wine_data['fixed acidity'] >= 0]
white_wine_data = white_wine_data[white_wine_data['fixed acidity'] >= 0]
red_wine_data = red_wine_data[red_wine_data['citric acid'] >= 0]
white_wine_data = white_wine_data[white_wine_data['citric acid'] >= 0]

# Удаление дубликатов
red_wine_data.drop_duplicates(inplace=True)
white_wine_data.drop_duplicates(inplace=True)

print("Размер после предобработки:")
print("Красное вино:", red_wine_data.shape)
print("Белое вино:", white_wine_data.shape)
```

```python
# Расчет дисперсии и стандартного отклонения параметра 'alcohol'
var_alcohol = red_wine_data['alcohol'].var()
std_alcohol = red_wine_data['alcohol'].std()

# Вывод результатов
print("Параметр 'alcohol' в красном вине:")
print("Дисперсия:", var_alcohol)
print("Стандартное отклонение:", std_alcohol)

var_alcohol = white_wine_data['alcohol'].var()
std_alcohol = white_wine_data['alcohol'].std()

# Вывод результатов
print("Параметр 'alcohol' в белом вине:")
print("Дисперсия:", var_alcohol)
print("Стандартное отклонение:", std_alcohol)
```

```python
# Выбор двух параметров для анализа линейной зависимости
param1 = 'alcohol'
param2 = 'pH'

# Расчет коэффициента корреляции Пирсона
correlation = red_wine_data[param1].corr(red_wine_data[param2])

# Вывод результата
print(f"Линейная зависимость между '{param1}' и '{param2}':")
print("Коэффициент корреляции:", correlation)
```

```python
import matplotlib.pyplot as plt

# Построение диаграммы рассеяния
plt.scatter(red_wine_data[param1], red_wine_data[param2])

# Настройка заголовка и меток осей
plt.title(f"Диаграмма рассеяния между '{param1}' и '{param2}'")
plt.xlabel(param1)
plt.ylabel(param2)

# Отображение диаграммы рассеяния
print('Красное вино')
plt.show()
```

```python
# Построение диаграммы рассеяния
plt.scatter(white_wine_data[param1], white_wine_data[param2])

# Настройка заголовка и меток осей
plt.title(f"Диаграмма рассеяния между '{param1}' и '{param2}'")
plt.xlabel(param1)
plt.ylabel(param2)

# Отображение диаграммы рассеяния
print('Белое вино')
plt.show()
```

```python
import seaborn as sns

param = 'alcohol'

# Построение гистограммы
sns.histplot(red_wine_data[param], kde=True)

# Настройка заголовка и меток осей
plt.title(f"Гистограмма параметра '{param}' для Красного вина")
plt.xlabel(param)
plt.ylabel("Частота")

# Отображение гистограммы
plt.show()
```

```python
# Построение гистограммы
sns.histplot(white_wine_data[param], kde=True)

# Настройка заголовка и меток осей
plt.title(f"Гистограмма параметра '{param}' для белого вина")
plt.xlabel(param)
plt.ylabel("Частота")

# Отображение гистограммы
plt.show()
```
